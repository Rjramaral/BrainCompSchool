Implemente funções em Python para aproximar a integral definida
$$
I = \int_a^b f(x)\,dx,
$$
usando:
i) a regra dos trapézios repetida/composta;
ii) a regra de Simpson repetida/composta.

a) Considere como parâmetros de entrada o intervalo $[a, b]$, a função $f$ e a quantidade de nós (pontos) que serão usados.

**Observação**: Você precisará implementar 2 funções.  Faça isso sem recorrer a funções de bibliotecas ou módulos de terceiros que implementem métodos de integração numérica.


```python
immport math

def trpezio(f,a,b,n)
h = (b-a)/n
soma_pontas = f(a) + f(b)
soma_loop = 0
for i in range(1,n):
  xi = a + i*h
  soma_loop += f(xi)
resultado = (h/2)*(soma_pontas + 2*soma_loop)
return resultado
```

Estruturando o código

A regra do trapézio para integração numérica é dada por:

$$\int_a^b f(x) dx \approx \frac{h}{2} \left[ f(x_0) + 2\sum_{i=1}^{n-1} f(x_i) + f(x_n) \right]$$

podemos perceber que há um loop onde os valores da função vão ser multiplicados por 2


inicialmente comecei o código pensando em uma função que vai receber os dados, pensei inicialmente em 4 parâmetros de entreda

f = função que queremos integrar
a = O limite inferior
b = O limite superior
n = O número de subintervalos

alem disso, a fórmula para calcular os pontos dentro do loop seria

xi = a + i*h

pela fórmula podemos perceber q as extremidades onde f(x0) e f(xn), ponto inicial e ponto final são usadas apenas uma vez, por isso chamei soma das pontas de f(a) + f(b)
